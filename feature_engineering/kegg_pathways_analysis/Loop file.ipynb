```python
pathwaynum = []
```

```python
d = {}
path_merge = {}
np.set_printoptions(suppress=True)
for i in range(len(list_of_pathways)):
    pathwaynum.append(list_of_pathways[i])
    d["pathdf" + str(i)] = pd.DataFrame({'Approved_Symbol':pathwaynum[i]}).dropna()
    d["pathdf" + str(i)] = d["pathdf" + str(i)].reindex(d["pathdf" + str(i)].index.drop(d["pathdf" + str(i)].index[0]))
    path_merge[str(i)] = pd.merge(d["pathdf" + str(i)],all_genes,on='Approved_Symbol')
    path_merge[str(i)].loc['mean'] = path_merge[str(i)].mean()
    path_merge[str(i)].loc['variance'] =  np.var(path_merge[str(i)])
```

```python
for j in range(len(list_of_pathways)):
    dffinal = dffinal.append(path_merge[str(j)].loc['mean',:])
    dffinal = dffinal.append(path_merge[str(j)].loc['variance',:])
```
